# 随机森林（Random Forest）

随机森林是一种**集成学习（Ensemble Learning）**算法。它的基本单元是**决策树**，每颗决策树都是一个分类器，那么对于一个输入样本，N颗树就会有N个分类结果，最终通过投票将得票次数最多的类别指定为最终的输出，这是一种**Bagging（套袋法）**思想。

随机森林相当于机器学习领域的多面手(Leatherman)，你几乎可以把任何东西扔进去，它基本上都是可供使用的。


## 决策树的生成
森林中的每棵树按照如下规则生成：

1. 如果训练集大小为$N$，对于每棵树而言，**随机且有放回地**从训练集中的抽取N个训练样本（称为**bootstrap采样方法**），作为该树的训练集，这样每棵树的训练集都是不同的，而且里面会包含重复的训练样本；

> **Bootstrap Sample**：有放回抽样  
> **为什么要随机抽样**：不进行随机抽样，每棵树的训练集都一样，那么最终训练出的树分类结果也是完全一样的  
> **为什么要有放回地抽样**：不放回，那所有树的训练集都不一样，每棵树都是"有偏见"的，结果差异会很大，无法投票  
> **两个随机性的引入对随机森林的分类性能至关重要**，由于它们的引入，使得随机森林不容易陷入过拟合，并且具有很好得抗噪能力（比如：对缺省值不敏感）

2. 如果每个样本的特征维度为$M$，指定一个常数$m<<M$，随机地从$M$个特征中选取$m$个特征子集，每次树进行分裂时，从这$m$个特征中选择最优的（特征选择可使用如：C4.5、ID3和CART等算法）；

3. 每棵树都尽最大程度的生长，并且没有剪枝过程；

### 决策树参数

随机森林分类效果（错误率）与两个因素有关：
- 森林中任意两棵树的相关性：**相关性越大，错误率越大**；
- 森林中每棵树的分类能力：**每棵树的分类能力越强，整个森林的错误率越低**。

减小特征选择个数$m$，树的相关性和分类能力也会相应的降低；增大$m$，两者也会随之增大。

所以关键问题是如何选择最优的$m$（或者是范围），这也是随机森林唯一的一个参数。

## 袋外错误率（oob error）
对于如何选择最优的$m$，主要依据计算袋外错误率oob error（out-of-bag error）。

随机森林有一个重要的优点就是，没有必要对它进行交叉验证或者用一个独立的测试集来获得误差的一个无偏估计。它可以在内部进行评估，也就是说在生成的过程中就可以对误差建立一个无偏估计。

对于每棵树而言（假设对于第k棵树），大约有1/3的训练实例没有参与第k棵树的生成，它们称为第k棵树的oob样本。
