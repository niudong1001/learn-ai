```python
import pandas as pd
```

### pandas数据类型
Pandas 基于两种数据类型，series 和 dataframe。  

series 是一种一维的数据类型，其中的每个元素都有各自的标签。如果你之前看过这个系列关于 Numpy 的推文，你可以把它当作一个由带标签的元素组成的 numpy 数组。标签可以是数字或者字符。

dataframe 是一个二维的、表格型的数据结构。Pandas 的 dataframe 可以储存许多不同类型的数据，并且每个轴都有标签。你可以把它当作一个 series 的字典。

```python
df = pd.read_csv("../sources/uk_rain_2014.csv", header=0)  # header 关键字告诉 Pandas 哪些是数据的列名
df.head(5)  # 查看前 x 行的数据
df.tail(5)  # 查看后 x 行的数据
```

```python
# 修改列名称
df.columns = ['water_year','rain_octsep', 'outflow_octsep',
              'rain_decfeb', 'outflow_decfeb', 'rain_junaug', 'outflow_junaug']
df.head(5)
```

```python
len(df)  # 有多少行数据
```

```python
pd.options.display.float_format = "{:,.3f}".format  # 格式化float显示3位小数点后
df.describe()  # 获得统计数据
```

```python
df["water_year"]  # 获得一个series，可以直接用df.water_year访问
```

```python
df.outflow_octsep < 1000  # 布尔过滤，返回dataframe
```

```python
df[(df.outflow_octsep<5000) & (df.outflow_octsep>4300)]
```

```python
df[df.water_year.str.startswith('199')]  # 字符串过滤
```

```python
df.iloc[30]  # iloc 只对数字型的标签有用。它会返回给定行的 series
```

```python
df = df.set_index(["water_year"])  # 用某一列索引，当你将一列设置为索引的时候，它就不再是数据的一部分了，
df.head(5)
```

```python
df.loc["1984/85"]  # 字符型索引
```

```python
df.sort_index(ascending=False).head(5) #inplace=True to apple the sorting in place
```

```python
df = df.reset_index("water_year")  # 恢复索引为数据
df.head(5)
```
